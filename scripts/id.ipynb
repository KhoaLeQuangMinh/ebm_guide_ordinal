```python
import pandas as pd
import numpy as np

# 1. Load the datasets
# Assuming DXSUM.csv and ADSL.csv are in the current directory
dxsum = pd.read_csv('/Users/khoale/Downloads/DXSUM.csv', low_memory=False)
adsl = pd.read_csv('/Users/khoale/Downloads/ADSL.csv', low_memory=False)

# 2. Define visit order and ranking
visit_order = ["sc", "bl", "m06", "m12", "m18", "m24", "m36", "m48", "m60", "m72"]
visit_rank_map = {visit: rank + 1 for rank, visit in enumerate(visit_order)}

# Filter DXSUM for relevant visits and add the rank column
dx_ranked = dxsum[dxsum['VISCODE2'].isin(visit_order)].copy()
dx_ranked['rank'] = dx_ranked['VISCODE2'].map(visit_rank_map)

# Map diagnoses to a numeric severity level
def map_diag_sev(diag):
    if diag == "CN":
        return 0
    elif diag == "MCI":
        return 1
    elif diag in ["AD", "Dementia"]:
        return 2
    else:
        return np.nan

dx_ranked['diag_sev'] = dx_ranked['DIAGNOSIS'].apply(map_diag_sev)

# 3. Identify Reverters
# A reverter is someone whose diagnosis severity DECREASES as the visit rank INCREASES.
dx_ranked_sorted = dx_ranked.sort_values(['PTID', 'rank'])
dx_ranked_sorted['prev_sev'] = dx_ranked_sorted.groupby('PTID')['diag_sev'].shift(1)

reverters = dx_ranked_sorted[
    (dx_ranked_sorted['diag_sev'].notna()) & 
    (dx_ranked_sorted['prev_sev'].notna()) & 
    (dx_ranked_sorted['diag_sev'] < dx_ranked_sorted['prev_sev'])
]['PTID'].unique()

print(f"Number of reverters identified: {len(reverters)}")

# 4. Identify MCI Start IDs (ADNI1, ADNI2)
mci_start_ids = dxsum[
    (dxsum['COLPROT'].isin(["ADNI1", "ADNI2"])) &
    (dxsum['VISCODE2'].isin(["sc", "bl"])) &
    (dxsum['DIAGNOSIS'] == "MCI")
]['PTID'].unique()

# 5. Extract pMCI (The Converters)
# Logic: Started as MCI, converted to Dementia/AD between m06 and m36, not a reverter
pmci_mask = (
    (dxsum['PTID'].isin(mci_start_ids)) &
    (dxsum['VISCODE2'].isin(["m06", "m12", "m18", "m24", "m36"])) &
    (~dxsum['PTID'].isin(reverters)) &
    (dxsum['DIAGNOSIS'].isin(["Dementia", "AD"]))
)
pMCI_RIDs = dxsum[pmci_mask]['PTID'].unique()

# 6. Extract sMCI (The Stable MCI)
# Logic: Started as MCI, NOT a converter, still MCI at m36, not a reverter
smci_mask = (
    (dxsum['PTID'].isin(mci_start_ids)) &
    (~dxsum['PTID'].isin(pMCI_RIDs)) &
    (dxsum['VISCODE2'] == "m36") &
    (dxsum['DIAGNOSIS'] == "MCI") &
    (~dxsum['PTID'].isin(reverters))
)
sMCI_RIDs = dxsum[smci_mask]['PTID'].unique()

print(f"Number of pMCI subjects: {len(pMCI_RIDs)}")
print(f"Number of sMCI subjects: {len(sMCI_RIDs)}")

# 7. Create the labels dataframe and merge features
pmci_df = pd.DataFrame({'PTID': pMCI_RIDs, 'Label': 'pMCI'})
smci_df = pd.DataFrame({'PTID': sMCI_RIDs, 'Label': 'sMCI'})
final_rid_list = pd.concat([pmci_df, smci_df], ignore_index=True)

# Extract SUBJID for merging (last 4 chars of PTID)
final_rid_list['SUBJID'] = final_rid_list['PTID'].str[-4:].astype(int)
adsl['SUBJID'] = adsl['SUBJID'].astype(int)

# Merge clinical/demographic features from ADSL
final_clinical_dataset = final_rid_list.merge(adsl, on='SUBJID', how='inner')

# Filter out current Dementia cases from the MCI cohort to ensure clean labels
final_clinical_dataset_removed = final_clinical_dataset[final_clinical_dataset['DX'] != 'DEM'].copy()

# 8. Balance the dataset with AD and CN subjects
n_mci = len(final_clinical_dataset_removed)
target_per_class = n_mci // 2

# Sample AD and CN from original ADSL
ad_data = adsl[adsl['DX'].isin(["DEM", "AD"])].sample(n=target_per_class, random_state=42).copy()
ad_data['Label'] = 'AD'

cn_data = adsl[adsl['DX'] == "CN"].sample(n=target_per_class, random_state=42).copy()
cn_data['Label'] = 'CN'

# Identify common columns to combine all groups
common_cols = list(set(ad_data.columns) & set(final_clinical_dataset_removed.columns))
if 'Label' in common_cols:
    common_cols.remove('Label')

final_balanced_dataset = pd.concat([
    final_clinical_dataset_removed[common_cols + ['Label']],
    ad_data[common_cols + ['Label']],
    cn_data[common_cols + ['Label']]
], ignore_index=True)

# 9. Final Join with PTID mapping from DXSUM
ptid_mapping = dxsum[['RID', 'PTID', 'VISCODE2', 'EXAMDATE']].drop_duplicates(subset=['RID']).copy()
ptid_mapping = ptid_mapping.rename(columns={'RID': 'SUBJID'})

final_balanced_with_ptid = final_balanced_dataset.merge(ptid_mapping, on='SUBJID', how='left', suffixes=('', '_mapped'))

# 10. Save results
final_balanced_with_ptid.to_csv('ADNI12_CN_sMCI_pMCI_AD.csv', index=False)
final_clinical_dataset_removed.to_csv('MCI_Cohort_with_Features.csv', index=False)
final_rid_list.to_csv('MCI_Conversion_List.csv', index=False)

print(f"Final dataset processed. Total subjects: {len(final_balanced_with_ptid)}")
print(final_balanced_with_ptid['Label'].value_counts())
```

```python
import pandas as pd
import numpy as np
import re

# 1. LOAD DATA
dxsum = pd.read_csv('/Users/khoale/Downloads/DXSUM_24Apr2026.csv', low_memory=False)
adsl = pd.read_csv('/Users/khoale/Downloads/ADSL.csv', low_memory=False)

# 2. PRE-PROCESSING
def extract_month(viscode):
    if viscode in ['bl', 'sc']: return 0
    match = re.match(r"m(\d+)", str(viscode))
    return int(match.group(1)) if match else None

dxsum['month'] = dxsum['VISCODE2'].apply(extract_month)

def map_diag_sev(diag):
    if diag == 1.0: return 0 # CN
    elif diag == 2.0: return 1 # MCI
    elif diag == 3.0: return 2 # AD
    return np.nan

dxsum['diag_sev'] = dxsum['DIAGNOSIS'].apply(map_diag_sev)

# Sort chronologically to track history
dxsum = dxsum.sort_values(['PTID', 'month'])

# 3. CALCULATE CLINICAL HISTORY
# Reverter Check
dxsum['prev_sev'] = dxsum.groupby('PTID')['diag_sev'].shift(1)
reverters = dxsum[dxsum['diag_sev'] < dxsum['prev_sev']]['PTID'].unique()

# History Stats
conversions_to_ad = dxsum[dxsum['DIAGNOSIS'] == 3.0].groupby('PTID')['month'].min()
max_followup = dxsum.groupby('PTID')['month'].max()
max_sev_reached = dxsum.groupby('PTID')['diag_sev'].max()

# 4. PREPARE THE STRICT BASELINE POOL (The Anchor)
# Restricted to ADNI1/2 and Month 0
baseline_pool = dxsum[(dxsum['PHASE'].isin(["ADNI1", "ADNI2"])) & (dxsum['month'] == 0)].copy()

# Fix the SC vs BL Priority (BL=0, SC=1) to ensure consistent date selection
baseline_pool['vis_priority'] = baseline_pool['VISCODE2'].map({'bl': 0, 'sc': 1}).fillna(2)
baseline_pool = baseline_pool.sort_values(['PTID', 'vis_priority', 'EXAMDATE'])

# 5. APPLY "ENTIRE ROW" LABELING LOGIC
is_reverter         = baseline_pool['PTID'].isin(reverters)
converts_to_ad      = baseline_pool['PTID'].isin(conversions_to_ad.index)
early_converter     = baseline_pool['PTID'].map(conversions_to_ad) <= 36
has_36m_followup    = baseline_pool['PTID'].map(max_followup) >= 36
stays_stable_cn     = baseline_pool['PTID'].map(max_sev_reached) == 0

conditions = [
    (baseline_pool['DIAGNOSIS'] == 2.0) & converts_to_ad & early_converter & ~is_reverter, # pMCI
    (baseline_pool['DIAGNOSIS'] == 2.0) & ~converts_to_ad & has_36m_followup & ~is_reverter, # sMCI
    (baseline_pool['DIAGNOSIS'] == 1.0) & stays_stable_cn & ~is_reverter, # CN
    (baseline_pool['DIAGNOSIS'] == 3.0) & ~is_reverter # AD
]

choices = ['pMCI', 'sMCI', 'CN', 'AD']
baseline_pool['Label'] = np.select(conditions, choices, default='Exclude')

# 6. DEDUPLICATE (Keep BL row if exists, else SC)
# Since we sorted by vis_priority, drop_duplicates will keep the "Right" row and date
final_labels_df = (
    baseline_pool[baseline_pool['Label'] != 'Exclude']
    .drop_duplicates('PTID', keep='first')
    .copy()
)

# 7. FINAL MERGE WITH ADSL
# Safely merge on SUBJID
final_labels_df['SUBJID_INT'] = final_labels_df['PTID'].str[-4:].astype(int)
adsl['SUBJID_INT'] = adsl['SUBJID'].astype(int)

final_dataset = final_labels_df.merge(adsl, on='SUBJID_INT', how='inner')

# 8. OUTPUT & LOGGING
print("="*30)
print("FINAL COHORT SELECTION SUMMARY")
print("="*30)
print(final_dataset['Label'].value_counts())
print(f"\nTotal Subjects with anchored EXAMDATE: {len(final_dataset)}")

final_dataset.to_csv('ADNI12_Stable_and_Converters.csv', index=False)
print("\nSuccess: Anchored 'ADNI12_Stable_and_Converters.csv' saved.")
```

```python
import pandas as pd
import numpy as np

# ─── PATHS ───────────────────────────────────────────────────────────────────
STABLE_PATH = "ADNI12_Stable_and_Converters.csv"
PET_PATH    = "/Users/khoale/Downloads/18FDG_PET_FLOURO_ADNI1_2_4_24_2026.csv"
MRI_PATH    = "/Users/khoale/Downloads/MRI_T1_3D_ADNI1_2_BL_PRE_4_24_2026.csv"
OUTPUT_PATH = "adni_pet_mri_cohort_2.csv"

# ─── CONSTRAINTS ─────────────────────────────────────────────────────────────
MAX_DATE_GAP_DAYS = 180   
AGE_MIN           = 55
AGE_MAX           = 91


# ─── EXPLICIT RANKING DEFINITIONS ────────────────────────────────────────────
# The index in these lists serves as the priority (0 is best)
MRI_DESC_PREF = [
    "MPR; GradWarp; B1 Correction; N3",
    "MPR-R; GradWarp; B1 Correction; N3",
    "MT1; GradWarp; N3m",
    "MPR; GradWarp; B1 Correction; N3; Scaled",
    "MPR-R; GradWarp; B1 Correction; N3; Scaled",
    "MPR; GradWarp; B1 Correction; N3; Scaled_2",
    "MPR-R; GradWarp; B1 Correction; N3; Scaled_2",
    "MPR; GradWarp; B1 Correction",
    "MPR-R; GradWarp; B1 Correction",
    "MPR; GradWarp; N3",
    "MPR-R; GradWarp; N3",
    "MPR; GradWarp; N3; Scaled",
    "MPR-R; GradWarp; N3; Scaled",
    "MPR; GradWarp; N3; Scaled_2",
    "MPR-R; GradWarp; N3; Scaled_2",
    "MPR; GradWarp",
    "MPR-R; GradWarp",
]
# Create a dictionary for O(1) lookup speed
MRI_DESC_RANK = {d: i for i, d in enumerate(MRI_DESC_PREF)}

PET_DESC_PREF = [
    "Coreg, Avg, Standardized Image and Voxel Size",
    "Coreg, Avg, Std Img and Vox Siz, Uniform Resolution",
    "Coreg, Avg, Std Img and Vox Siz, Uniform 6mm Res",
    "Co-registered, Averaged",
    "Co-registered Dynamic",
]
PET_DESC_RANK = {d: i for i, d in enumerate(PET_DESC_PREF)}

def rank_mri_quality(desc):
    # Returns the index if found, otherwise 999 (rejected)
    return MRI_DESC_RANK.get(desc, 999)

def rank_pet_quality(desc):
    return PET_DESC_RANK.get(desc, 999)

# # ─── RANKING FUNCTIONS ───────────────────────────────────────────────────────
# def rank_mri_quality(description):
#     d = str(description).upper()
#     # 1. Check for the GradWarp tag (Gold Standard for GE/Siemens)
#     has_gradwarp = "GRADWARP" in d
    
#     # 2. Check if it's a Philips scan (which often lacks the tag but is pre-corrected)
#     # Common Philips structural tags: "SENSE", "WIP", "PHILIPS"
#     is_philips = any(kw in d for kw in ["PHILIPS", "SENSE"]) 

#     # GATEKEEPER: Reject if it's not GradWarped AND not a known Philips variant
#     if not (has_gradwarp or is_philips):
#         return 999
#     if "SCALED" in d and "N3" in d: return 1
#     elif "N3" in d or "B1" in d: return 2
#     elif any(kw in d for kw in ["MPRAGE", "MPR", "T1", "SPGR", "MT1"]): return 3
#     return 999

# def rank_pet_quality(description):
#     d = str(description).upper()
#     if "STANDARDIZED" in d: return 1
#     elif "UNIFORM" in d: return 2
#     elif any(kw in d for kw in ["COREGESTERED", "COREG", "AVG", "AVERAGED"]): return 3
#     return 999

# ─────────────────────────────────────────────────────────────────────────────
# 1. LOAD DATA
# ─────────────────────────────────────────────────────────────────────────────
print("="*60)
print("STEP 1: LOADING RAW DATA")
print("="*60)
stable = pd.read_csv(STABLE_PATH)
pet    = pd.read_csv(PET_PATH)
mri    = pd.read_csv(MRI_PATH)

# Ensure dates are datetime objects
pet["Acq Date"] = pd.to_datetime(pet["Acq Date"])
mri["Acq Date"] = pd.to_datetime(mri["Acq Date"])
# Assuming EXAMDATE is the diagnosis date in your stable file
stable["EXAMDATE"] = pd.to_datetime(stable["EXAMDATE"]) 

print(f"  [Input] Subjects in Master List (Stable/Conv) : {len(stable)}")
print(f"  [Input] Total Raw PET rows                    : {len(pet)}")
print(f"  [Input] Total Raw MRI rows                    : {len(mri)}")

# ─────────────────────────────────────────────────────────────────────────────
# 2. INITIAL FILTER (PTID & AGE)
# ─────────────────────────────────────────────────────────────────────────────
print("\n" + "="*60)
print("STEP 2: PRE-FILTERING (PTID & AGE)")
print("="*60)

stable_ids = set(stable["PTID"])

# Filter PET
pet_valid = pet[pet["Subject"].isin(stable_ids) & pet["Age"].between(AGE_MIN, AGE_MAX)].copy()
pet_valid["PET_Tier"] = pet_valid["Description"].apply(rank_pet_quality)
pet_valid = pet_valid[pet_valid["PET_Tier"] < 999]

# Filter MRI
mri_valid = mri[mri["Subject"].isin(stable_ids) & mri["Age"].between(AGE_MIN, AGE_MAX)].copy()
mri_valid["MRI_Tier"] = mri_valid["Description"].apply(rank_mri_quality)
mri_valid = mri_valid[mri_valid["MRI_Tier"] < 999]

print(f"  [PET] Usable candidate rows (Correct Age/Desc): {len(pet_valid)}")
print(f"  [MRI] Usable candidate rows (Correct Age/Desc): {len(mri_valid)}")

# ─────────────────────────────────────────────────────────────────────────────
# 3. CREATE POTENTIAL PAIRS (THE "MASTER MERGE")
# ─────────────────────────────────────────────────────────────────────────────
print("\n" + "="*60)
print("STEP 3: CROSS-MATCHING ALL POSSIBLE PAIRS")
print("="*60)

# Merge PET and MRI on Subject ID to get every possible combo
# (One subject might have 3 PETs and 2 MRIs = 6 possible pairs)
pairs = pet_valid.merge(mri_valid, on="Subject", suffixes=('_PET', '_MRI'))

print(f"  Total potential PET-MRI combinations created: {len(pairs)}")

# Add the Diagnosis Date from the stable file to anchor the pairs
pairs = pairs.merge(stable[['PTID', 'EXAMDATE', 'Label']], left_on='Subject', right_on='PTID')

# ─────────────────────────────────────────────────────────────────────────────
# 4. THREE-WAY GAP FILTERING
# ─────────────────────────────────────────────────────────────────────────────
print("\n" + "="*60)
print("STEP 4: APPLYING THREE-WAY DATE CONSTRAINTS")
print("="*60)

# Calculate gaps
pairs["Gap_MRI_PET"] = (pairs["Acq Date_MRI"] - pairs["Acq Date_PET"]).dt.days.abs()
pairs["Gap_MRI_DX"]  = (pairs["Acq Date_MRI"] - pairs["EXAMDATE"]).dt.days.abs()
pairs["Gap_PET_DX"]  = (pairs["Acq Date_PET"] - pairs["EXAMDATE"]).dt.days.abs()

# Filter: All three gaps must be within the threshold
valid_pairs = pairs[
    (pairs["Gap_MRI_PET"] <= MAX_DATE_GAP_DAYS) &
    (pairs["Gap_MRI_DX"]  <= MAX_DATE_GAP_DAYS) &
    (pairs["Gap_PET_DX"]  <= MAX_DATE_GAP_DAYS)
].copy()

dropped_gap = len(pairs["Subject"].unique()) - len(valid_pairs["Subject"].unique())
print(f"  Subjects dropped (No pairs within {MAX_DATE_GAP_DAYS}d of each other/DX): {dropped_gap}")
print(f"  Remaining valid pairs: {len(valid_pairs)} across {valid_pairs['Subject'].nunique()} subjects")

# ─────────────────────────────────────────────────────────────────────────────
# 5. SELECT BEST PAIR (QUALITY FIRST, THEN PROXIMITY)
# ─────────────────────────────────────────────────────────────────────────────
print("\n" + "="*60)
print("STEP 5: SELECTING BEST QUALITY PAIR PER SUBJECT")
print("="*60)

# Priority Logic:
# 1. Best MRI Tier (1 is best)
# 2. Best PET Tier (1 is best)
# 3. Closest date to Diagnosis (MRI-DX gap)
# 4. Smallest MRI-PET gap
final_candidates = valid_pairs.sort_values(
    by=["Subject", "MRI_Tier", "PET_Tier", "Gap_MRI_DX", "Gap_MRI_PET"]
)

# Deduplicate to keep only the absolute best pair per subject
final = final_candidates.drop_duplicates(subset="Subject", keep="first").copy()

print(f"  [Success] Final Cohort Size: {len(final)} subjects")

# ─────────────────────────────────────────────────────────────────────────────
# 6. CLEANUP & SAVE
# ─────────────────────────────────────────────────────────────────────────────
# Rename columns for the final report
final = final.rename(columns={
    "Subject": "PTID",
    "Image Data ID_MRI": "MRI_ImageID",
    "Description_MRI": "MRI_Description",
    "Acq Date_MRI": "MRI_AcqDate",
    "Image Data ID_PET": "PET_ImageID",
    "Description_PET": "PET_Description",
    "Acq Date_PET": "PET_AcqDate"
})

# Final column selection
output_cols = [
    "PTID", "Label", "EXAMDATE", 
    "MRI_ImageID", "MRI_Description", "MRI_AcqDate", "MRI_Tier",
    "PET_ImageID", "PET_Description", "PET_AcqDate", "PET_Tier",
    "Gap_MRI_PET", "Gap_MRI_DX", "Gap_PET_DX"
]
final[output_cols].to_csv(OUTPUT_PATH, index=False)

print("\n--- FINAL SUMMARY ---")
print(final["Label"].value_counts().to_string())
print("\n--- QUALITY BREAKDOWN ---")
print(f"  Pairs with Tier 1 MRI: {len(final[final['MRI_Tier']==1])}")
print(f"  Pairs with Tier 1 PET: {len(final[final['PET_Tier']==1])}")
print(f"  Average MRI-PET Gap:   {final['Gap_MRI_PET'].mean():.1f} days")
print(f"✓ Saved: {OUTPUT_PATH}")
```
